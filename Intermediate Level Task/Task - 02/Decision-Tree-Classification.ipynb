## Decision Tree Classification Algorithm

A Decision tree is a tree-like structure that represents a set of decisions and their possible consequences. Each node in the tree represents a decision, and each branch represents an outcome of that decision. The leaves of the tree represent the final decisions or predictions.

Decision trees are created by recursively partitioning the data into smaller and smaller subsets. At each partition, the data is split based on a specific feature, and the split is made in a way that maximizes the information gain.

### Importing Required Libraries

```python
import pandas as pd
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

### Reading the csv file

```python
csv_file = pd.read_csv('Iris.csv')
csv_file
```

### Dropped the unnecessary column (Id)

```python
csv_file = csv_file.drop('Id', axis = 'columns')
csv_file
```

### Plotting the Species Column of the dataset

```python
plt.figure()
plt.plot(csv_file['Species'],label='Species')
plt.legend()
plt.show()
```

### Getting a summarized report of iris dataset

```python
csv_file.info()
```

### 1. Splitting the data into train and test set
### 2. Converting each set into a Dataframe
### 3. Fitting the Splitted data into Decision Tree Classifier and
### 4. Predicting the score of the model

```python
x_train, x_test, y_train, y_test = train_test_split(csv_file[['SepalLengthCm','SepalWidthCm','PetalLengthCm', 
                                                              'PetalWidthCm']], csv_file[['Species']], 
test_size = 0.25)

x_train = pd.DataFrame(x_train)
x_test = pd.DataFrame(x_test)
y_train = pd.DataFrame(y_train)
y_test = pd.DataFrame(y_test)

model = tree.DecisionTreeClassifier()
model.fit(x_train, y_train)

model.score(x_test, y_test)
```

### Predicting the species of the model with the help of LabelEncoder()

```python
model.predict([[5.9,3.0,5.1,1.8]])
```

### Plotting the 'Decision Tree Classification final output'

```python
plt.figure(figsize=(10,10))
tree.plot_tree(model, filled=True, fontsize=7)
plt.savefig('decision_tree.png')
```
